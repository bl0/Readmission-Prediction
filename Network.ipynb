```python
data = pd.read_csv('./data/data1218.csv')
```

```python
numpy_matrix = data.as_matrix()
```

```python
numpy_matrix = numpy_matrix.astype('float32')
```

```python
np.random.shuffle(numpy_matrix)
```

```python
numpy_matrix.shape
```

```python
train_data = numpy_matrix[:80000, 1:-1]
train_label = numpy_matrix[:80000, -1].astype('int32')
```

```python
print(train_data.shape)
print(train_label.shape)
```

```python
test_data = numpy_matrix[80000:, 1:-1]
test_label = numpy_matrix[80000:, -1].astype('int32')
```

```python
print(test_data.shape)
print(test_label.shape)
```

```python
train_data = mx.nd.array(train_data)
train_label = mx.nd.array(train_label)
test_data = mx.nd.array(test_data)
test_label = mx.nd.array(test_label)
```

```python
batch_size = 100
ctx = mx.cpu()
```

```python
train_iter = mx.io.NDArrayIter(data=train_data, label=train_label, batch_size=batch_size, shuffle=True)
test_iter = mx.io.NDArrayIter(data=test_data, label=test_label)
```

```python
def NaiveNet():
    data = mx.symbol.var(name='data')
    out = mx.symbol.FullyConnected(data=data, num_hidden=512, name='FC1')
    out = mx.symbol.Activation(data=out, act_type='relu')
    out = mx.symbol.FullyConnected(data=out, num_hidden=256, name='FC2')
    out = mx.symbol.Activation(data=out, act_type='relu')
    out = mx.symbol.FullyConnected(data=out, num_hidden=3, name='Out')
    out = mx.symbol.SoftmaxOutput(data=out, name='softmax')
    return out
```

```python
NNet = NaiveNet()
```

```python
Model = mx.mod.Module(symbol=NNet, context=ctx)
```

```python
Model.fit(train_iter, 
          eval_data=test_iter, 
          optimizer='sgd', 
          optimizer_params={'learning_rate':0.1},
          eval_metric='acc',
          batch_end_callback=mx.callback.Speedometer(batch_size, 100),
          num_epoch=10)
```
